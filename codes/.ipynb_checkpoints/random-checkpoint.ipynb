```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('advertising.csv')
df.head()
```

```python
# Remove features with string (non-numeric) values:
to_drop=["Ad Topic Line","City","Country","Timestamp"]
df.drop(columns=to_drop, inplace=True)
```

```python
df
```

```python
X_train = df.drop(columns = 'Clicked on Ad')
y_train = df['Clicked on Ad']
```

```python
# Scaling features to u = 0, sigma = 1
s = StandardScaler()
X_train_normalized = s.fit_transform(X_train)
```

```python
linear_model = SVC(kernel = 'linear')
poly_model = SVC(kernel = 'poly', degree = 2)
rbf_model = SVC(kernel = 'rbf')
```

```python
### Part A (Accuracies)

#linear
linear_model.fit(X_train_normalized,y_train)
y_pred = linear_model.predict(X_train_normalized)
linear_acc = accuracy_score(y_pred,y_train)

#poly
poly_model.fit(X_train_normalized,y_train)
y_pred = poly_model.predict(X_train_normalized)
poly_acc = accuracy_score(y_pred,y_train)

#rbf
rbf_model.fit(X_train_normalized,y_train)
y_pred = rbf_model.predict(X_train_normalized)
rbf_acc = accuracy_score(y_pred,y_train)

print("Accuracies:\nLinear: %.3f" %linear_acc, "\nPolynomial (degree 2): %.3f" %poly_acc, "\nRBF: %.3f" %rbf_acc)
```

```python
### Part B (Number of Support Vectors)

#linear
linear_model.fit(X_train_normalized,y_train)
linear_n = len(linear_model.support_vectors_)

#poly
poly_model.fit(X_train_normalized,y_train)
poly_n = len(poly_model.support_vectors_)

#rbf
rbf_model.fit(X_train_normalized,y_train)
rbf_n = len(rbf_model.support_vectors_)

print("No. of Support Vectors:\nLinear: %d" %linear_n, "\nPolynomial (degree 2): %d" %poly_n, "\nRBF: %d" %rbf_n)
```

```python
c = [10**i for i in range(-2,4)]
linear_n = np.zeros(len(c), dtype = np.int64)
poly_n = np.zeros(len(c), dtype = np.int64)
rbf_n = np.zeros(len(c), dtype = np.int64)

for i in range(len(c)):
    #linear
    linear_model.C = c[i]
    linear_model.fit(X_train_normalized,y_train)
    linear_n[i] = len(linear_model.support_vectors_)

    #poly
    poly_model.C = c[i]
    poly_model.fit(X_train_normalized,y_train)
    poly_n[i] = len(poly_model.support_vectors_)

    #rbf
    rbf_model.C = c[i]
    rbf_model.fit(X_train_normalized,y_train)
    rbf_n[i] = len(rbf_model.support_vectors_)
    
Result = {'C': c, 'Linear': linear_n, 'Polynomial': poly_n, 'RBF': rbf_n}
Result = pd.DataFrame(Result)
Result
```
